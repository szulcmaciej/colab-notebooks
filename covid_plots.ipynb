```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv')
```

```
df.head()
```

```
date_columns = [column for column in df.columns if '/20' in str(column)]
```

```
date_columns
```

```
df[df['Country/Region'] == 'Poland'].columns
```

```
country_agg = df.drop(['Lat', 'Long'], axis='columns').groupby('Country/Region').sum().reset_index()
# country_case_matrix = country_agg.iloc[:, 1:].values

# country_case_matrix = country_agg[country_agg['Country/Region'].isin(['Germany', 'France', 'Spain', 'Italy'])].iloc[:, 1:].values
```

```
plot_countries = ['Germany', 'France', 'Spain', 'Italy', 'Poland', 'US']


filtered_df = country_agg[country_agg['Country/Region'].isin(plot_countries)]

plt.figure(figsize=(17,10))
for i in range(len(filtered_df)):
  plt.plot(filtered_df.iloc[i, 1:])
  plt.grid(True)
  # plt.yscale('log')

plt.legend(filtered_df['Country/Region']);
```
